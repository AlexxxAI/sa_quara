```python
import pandas as pd

pred_file = "/Users/abubakr/zypl/scoring_rnd_my/Quara_new/project/predictions/jarir_sample101_predictions.xlsx"
data_file = "/Users/abubakr/zypl/scoring_rnd_my/Quara_new/project/data/jarir_sample101.xlsx"

pred_df = pd.read_excel(pred_file)
data_df = pd.read_excel(data_file)

# Compare model outputs
equal_rows = pred_df['proba'] == data_df['model_pred']
print("Rows with equal probabilities:\n", equal_rows.value_counts())

# Compute approval rate for threshold = 0.2
threshold = 0.1
approval_rate = (pred_df['proba'] <= threshold).mean()

print(f"\nApproval rate at threshold {threshold}: {approval_rate:.2%}")
```
